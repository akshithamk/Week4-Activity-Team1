#### Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

#### Reading file into dataframe

```python
data=pd.read_excel("Telco-Customer-Churn.xlsx")
data.head()
```

### 1) Compare churn count with respect to gender. 

Plotting churn count for each gender using count plot

```python
sns.countplot(x ='gender', hue='Churn', data=data)
plt.title('Churn count wrt to Gender');
```

### 2) Find out how many female senior citizens there in the dataset 

#### Method 1

Filtering all females to data 1 and then again filtering result to senior citizens

```python
data1=data[data['gender']=='Female']
data2=data1[data1['SeniorCitizen']==1]
```

Counting all female senior citizens

```python
data2.SeniorCitizen.count()
```

Filtering out senior citizens to dataw then plotting a bar graph based on gender value count

```python
data2=data[data['SeniorCitizen'] == 1 ] 
data2['gender'].value_counts().plot(kind="bar")
plt.title("Senior Citizen count", y=1.02)
plt.ylim(550,600);
```

### 4) Find out which contract preffered by the senior citizen.

#### Plot 1

```python
sns.countplot(data=data,x='Contract',hue='SeniorCitizen')
plt.title('Contract preffered by Senior Citizen')
plt.show() 
```

#### Plot 2

```python
data3=data[data['SeniorCitizen'] == 1 ]
data3["Contract"].value_counts()
data3["Contract"].value_counts().plot(kind='pie')
plt.legend(loc='best');
```

It can be inferred from both kinds of plot that senior citizens prefer Month-to-month contract over One year and Two year contracts
